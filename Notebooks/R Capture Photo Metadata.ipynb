# SETUP - import packages

```r
#INSTALL IF REQUIRED
#install.packages('exiftoolr')
#install_exiftool()

#IMPORT LIBRARIES
library(exiftoolr)
library(tibble)
```

# Construct list of files to query

```r
#construct list of files
image_files <- dir(path = 'G:/My Drive/9.1 Photos/Camera', full.names = TRUE, recursive = TRUE)
image_files
```

# Read image metadata

```r
#read image metadata
exifinfo <- exif_read(image_files)
tibble(exifinfo)
```

# See what fields are avaiable

```r
#seeing what fields are avaiable
#could add columns to dataframe for video
names(exifinfo)
```

# Constructing dataframe

```r
#constructing dataframe
df <- exif_read(image_files, tags = c("filename","make","model","image size", "GPSLatitude","GPSLongitude", "XResolution" , "YResolution", 'FileSize','FileTypeExtension', "FileCreateDate", "FileModifyDate", "ImageSize", "Megapixels"   ))
tibble(df)
```

```r
df
```

# Export results

```r
write.csv(df, ".\\r\\g drive file extract photos - my camera.csv")
```
